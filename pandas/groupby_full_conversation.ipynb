```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'key': ['a','b','c'] * 4,
    'A': np.arange(12),
    'B': [1,2,3] * 4,
})

# Group and sum
result = (
    df.groupby('key', as_index=False)
      .agg(sum_A=('A', 'sum'))
)

# Add percentage column
total_sum = result['sum_A'].sum()
result['percentage'] = result['sum_A'] / total_sum * 100


# Create TOTAL row
total_row = pd.DataFrame({
    'key': ['Total'],
    'sum_A': [total_sum],
    'percentage': [100.0]
})

# Append TOTAL row
result = pd.concat([result, total_row], ignore_index=True)

result
```

### Explanation / Notation
- `groupby('key')` : Groups rows based on the 'key' column.
- `sum()` : Aggregates the sum for column 'A' per group.
- `percentage` : Each group’s sum as a percentage of total sum.
- `TOTAL row` : Added manually for reporting purposes.
- `sort_values('sum_A', ascending=False)` : Sorts groups from highest to lowest sum.
- The final table keeps Total pinned at the bottom for readability.
