```python
import pandas as pd

# Load the data
df = pd.read_csv('canadian_financial_ratings.csv')

# Define a function to categorize scores into 'Promoter', 'Passive', or 'Detractor'
def categorize_nps(score):
    if score >= 9:
        return 'Promoter'
    elif score >= 7:
        return 'Passive'
    else:
        return 'Detractor'

# Apply the categorization function to create a new 'category' column
df['category'] = df['rating'].apply(categorize_nps)

# Group the data by brand/group and calculate the NPS for each
def calculate_nps(group):
    total_responses = group['category'].count()
    promoters = group[group['category'] == 'Promoter']['category'].count()
    detractors = group[group['category'] == 'Detractor']['category'].count()
    
    # Calculate percentages
    percent_promoters = (promoters / total_responses) * 100
    percent_detractors = (detractors / total_responses) * 100
    
    # NPS formula: % Promoters - % Detractors
    nps_score = percent_promoters - percent_detractors
    ##return round(nps_score, 2) # Round to 2 decimal places
    return round(nps_score) # Round to 2 decimal places

# Apply the NPS calculation function per group
nps_per_group = df.groupby('financial_institution').apply(calculate_nps).rename('NPS Score')

print(nps_per_group)
```

```python
import pandas as pd
import numpy as np

def calculate_nps(scores):
    """
    Calculates the Net Promoter Score (NPS) from a series of scores (0-10).
    NPS = % Promoters (9-10) - % Detractors (0-6)
    """
    total = len(scores)
    if total == 0:
        return 0
    promoters = np.sum(scores >= 9)
    detractors = np.sum(scores <= 6)
    nps_score = ((promoters - detractors) / total) * 100
    return round(nps_score) # NPS is typically presented as a whole number

# Sample data simulation (replace with your actual data loading)
df = pd.read_csv('canadian_financial_ratings.csv')

# Calculate NPS per group
nps_per_group = df.groupby('financial_institution')['rating'].agg(calculate_nps)

print(nps_per_group)


# Define a function to categorize scores into 'Promoter', 'Passive', or 'Detractor'
def categorize_nps(score):
    if score >= 9:
        return 'Promoter'
    elif score >= 7:
        return 'Passive'
    else:
        return 'Detractor'

# Apply the categorization function to create a new 'category' column
df['category'] = df['rating'].apply(categorize_nps)

df.groupby(['financial_institution','category']).size()
```
