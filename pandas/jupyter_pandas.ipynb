```python
df_group['Sum_per_cat'] = df_group.groupby('Category')['Price'].transform('sum')

##df_group['Ptc_Cat_Price'] = (df_group['Price'].sum() / df_group['Sum_per_cat'])
##df_group['Ptc_Cat_Price'] = df_group['Price'].sum() / df_group['Sum_per_cat']

df_group['Ptc_Cat_Price'] =  (df_group.groupby('Category')['Price'].transform('sum') / df_group['Price'].sum()) * 100
```

```python
df_group.groupby('Category')['Sum_per_cat'].first()
##df_group.groupby('Category')['Price'].transform('sum')


##df_group.groupby(['Category','Ptc_Cat_Price']).value_counts()

##df_gcat = df.groupby('Category')

##print(type(df))
##print(type(df_gcat))

df_group.groupby('Category')['Ptc_Cat_Price'].first().round().to_frame('ptc')
##df.groupby('Category').get_group('Furniture').


df_group.groupby('Category')[['Price', 'Ptc_Cat_Price']].agg(['sum', 'first'])

##https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.DataFrameGroupBy.aggregate.html
##df.groupby('A').agg({'B': ['min', 'max'], 'C': 'sum'})
df.groupby('Category').agg({'Price':'sum', 'Ptc_Cat_Price': 'first'}).round()

####df.groupby("A").agg(
####    b_min=pd.NamedAgg(column="B", aggfunc="min"),
####    c_sum=pd.NamedAgg(column="C", aggfunc="sum")
####)

df_group.groupby('Category').agg(
    Count=pd.NamedAgg(column="Price", aggfunc="sum"),
    Percent=pd.NamedAgg(column="Ptc_Cat_Price", aggfunc="first")
).round(0).astype('Int64')

combined_results = pd.concat([df_group.groupby('Category').agg({'Price':'sum'}).round().astype('Int64'), df_group.groupby('Category').agg({'Ptc_Cat_Price':'first'}).round().astype(str).replace(r'\.0', '%', regex=True)], axis=1)
combined_results.columns = ['Count', 'Percent']

print(combined_results.columns)
aa = df_group.groupby('Category').agg(
    Count=pd.NamedAgg(column="Price", aggfunc="sum")).round().astype('Int64')

bb = df_group.groupby('Category').agg(
   
    Percent=pd.NamedAgg(column="Ptc_Cat_Price", aggfunc="first")
).round()

combined_results = pd.concat([aa,bb], axis=1)

print(combined_results)
```
